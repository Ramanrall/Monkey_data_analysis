```python
import pandas as pd
import os
```

```python
pwd = os.getcwd()
```

```python
dataset = pd.read_excel(pwd + "/Data - Survey Monkey Output Edited.xlsx", sheet_name="Edited_data")
dataset
```

```python
dataset_modified = dataset.copy()
dataset_modified
```

```python
dataset_modified.columns
```

```python
columns_to_drop = ['Start Date', 'End Date', 'Email Address','First Name', 'Last Name', 'Custom Data 1 - Response']
columns_to_drop
```

```python
dataset_modified = dataset_modified.drop(columns= columns_to_drop)
dataset_modified
```

```python
id_vars = dataset_modified.columns[0 : 8]
value_vars = dataset_modified.columns[8 : ]
```

```python
data_melted = dataset_modified.melt(id_vars = id_vars, value_vars = value_vars, var_name="Question + subqustions", value_name= "Answers")
data_melted
```

```python
questions_import = pd.read_excel(pwd + "/Data - Survey Monkey Output Edited.xlsx", sheet_name="Question" )
questions_import
```

```python
questions = questions_import.copy()
questions.drop(columns=["Raw Question", "Raw subquestion", "subquestion"], inplace=True)
questions
```

```python
dataset_merge = pd.merge(left= data_melted, right=questions, how="right", left_on="Question + subqustions", right_on="question + subquestion")
dataset_merge
```

```python
respondents = dataset_merge[dataset_merge["Answers"].notna()]
respondents = respondents.groupby("Question")["Respondent ID"].nunique().reset_index()
respondents.rename(columns={"Respondent ID" : "Respondents"}, inplace=True)
respondents
```

```python
dataset_merge_two = pd.merge(left= dataset_merge, right=respondents, how="right", left_on="Question", right_on="Question")
dataset_merge_two
```

```python
same_answer = dataset_merge_two
same_answer = same_answer.groupby(["Question + subqustions", "Answers"]) ["Respondent ID"].nunique().reset_index()
same_answer.rename(columns= {"Respondent ID" : "Same Answer"}, inplace=True)
same_answer
```

```python
dataset_merge_three = pd.merge(left= dataset_merge_two, right=same_answer, how="right", left_on=["Question + subqustions", "Answers"], right_on=["Question + subqustions", "Answers"])
dataset_merge_three["Same Answer"].fillna(0)
print("Orignal Data", len(dataset_merge_two))
print("Merge Data", len(dataset_merge_three))
dataset_merge_three
```

```python
output = dataset_merge_three.copy()
```

```python
output.rename(columns={"Identify which division you work in. - Other (please specify)": "Primary Division",
"Identify which division you work in. - Response": "Secondary Division",
"Which of the following best describes your position level? - Response": "Level",
"Which generation are you apart of? - Response": "Generation",
"Please select the gender in which you identify. - Response": "Gender",
"Which duration range best aligns with your tenure at your company? - Response": "Tenure",
"Which of the following best describes your employment type? - Response": "Employment Type"}, inplace=True)
output
```

```python
output.to_excel(pwd + "/Final_Output.xlsx", index=False)
```
